```python
import flopy
import os
import sys

modelname = 'test_write_input'
sw_exe = '/Users/ianpg/Documents/ProjectsLocal/SyntheticSWI/bin/modflow_macOS/swt_v4'
fname = os.path.join('/Users/ianpg/Documents/ProjectsLocal/DelawareSGD/work', modelname, modelname + '.nam')
model_ws = os.path.join('/Users/ianpg/Documents/ProjectsLocal/DelawareSGD/work', modelname)

m = flopy.seawat.Seawat.load(fname,exe_name = sw_exe, model_ws = model_ws)
v = m.run_model(silent=False, report=True)
```

```python
#Get custom fuctions for post-processing
sys.path.append('Users/ianpg/Documents/ProjectsLocal/DelawareSGD/')
import utils
import numpy as np
#Get flux at ocean boundary
ocean_col = [30,69]
ocean_col_vec = np.arange(ocean_col[0],ocean_col[1]+1)
ocean_outflow = utils.get_ocean_outflow_chd(m,ocean_col)
```

```python
#Print out coordinates and flux to text file
fout= open(os.path.join(model_ws,"flux.pts"),"w")
fout.write('Values are zero-based \n')
fout.write('{:14s} {:4s} {:4s} {:4s} \n'.format("flux", "lay","row", "col"))
for i in range(len(ocean_outflow)):
     fout.write('{:14.8f} {:4d} {:4d} {:4d}\n'.format(ocean_outflow[i], 0,0,ocean_col_vec[i]))
fout.close()
```

```python
#Write an instruction file
finst = open(os.path.join(model_ws,"flux.ins"),"w")
finst.write('pif #\n')
finst.write('#flux#\n')
for i in range(len(ocean_outflow)):
     finst.write('l1 w !{:s}!\n'.format('flux_' + str(i)))
finst.close()
```
